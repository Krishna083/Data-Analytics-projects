```python
# Display the percentage distribution of the data in each year using the pie chart.
plt.figure(figsize = (7,7))
plt.pie(df.year.value_counts(),autopct = '%.1f%%',labels=df.year.value_counts().index)
plt.show()
```

```python
# Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings
sns.displot(df['priceperlb'],bins = 20, kde = True )
```

# Observation - Most of the data lies between 0.5 - 1.7 (priceperlb). The data is skewed towards right side this means mean will shift to right side. As per the data, sales was very low after 2.6 priceperlb.

```python
# Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

numcol = df['numcol']
prodvalue = df['prodvalue']
plt.xlabel("numcol")
plt.ylabel("prodvalue")
plt.scatter(numcol, prodvalue, color = 'red')
plt.show()

```

# Observation - There is a  positive relationship between variables numcol and prodvalue. If there is change in one variable the other variable also changes. There are some outliers in variable numcol.

```python
#  Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings
plt.figure(figsize = (7,7))
sns.boxplot(data = df, x = 'year', y = 'prodvalue')
plt.show
```

# Observation - As per the data there are many outliers in it. There is positive relation between both the variables but not too strong. The prodvalue in initial years decreases and it started increasing in year 2002,3,4. Again there was decline in prodvalue in year 2005,6,7 and afterwards there was a good growth in later years. In the year 2010,11,12 the prodvalue was highest.

```python
# Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')
plt.figure()
sns.pairplot(df[["numcol", "yieldpercol", "totalprod", "prodvalue","year"]], hue = "year", kind = "scatter", palette = "husl")
plt.show()
```

# Observation - The plots that they are have positive uptrend relationship with each other. The attributes numcol, totalprod, prodvalue clearly indicates an extreme long tail on the right. Presence of higher values. Yieldpercol should probably have a normal distribution. The constraints on the data collection may have skewed the distribution.
